```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+","",text) #link remove
    text = re.sub(r"<.*?>","",text) # html tag remove
    text = re.sub(r"[^a-z\s]","",text) #punctuations and digit remove
    text = re.sub(r"\s+"," ",text) #extra space remove
    return text.strip() #remove very first and last space of the sentence if any exist

df['clean_review'] = df['review'].apply(clean_text)
```
